```python
import pandas as pd

df=pd.read_csv("../YSE_finder/Shane_2023-01-30_last.txt",sep="\s+")
df.columns = ['name', 'ra_h', 'ra_m', 'ra_s', 'dec_d', 'dec_m',
       'dec_s', 'equinox', 'mag','pr','{comment', '*}']
```

```python
df2=df.sort_values(by='ra_h')
df2.to_csv("test.txt",sep=" ", index=False)
output = df2.to_string(formatters={
    'ra_h': '{:0>2}'.format,
    'ra_m': '{:0>2}'.format,
    'ra_s': '{:05.2f}'.format,
    'dec_d': '{:02.0f}'.format,
    'dec_m': '{:0>2}'.format,
    'dec_s': '{:05.2f}'.format,
}, index=False)
print(output)
```
